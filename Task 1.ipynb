## Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable,such as the distribution of ages or genders in a population.

```python
import zipfile
import pandas as pd
import matplotlib.pyplot as plt
import os

# Step 1: Unzip the dataset
zip_path = "API_SP.POP.TOTL_DS2_en_csv_v2_1237901.zip"
extract_path = "population_data"

with zipfile.ZipFile(zip_path, "r") as zip_ref:
    zip_ref.extractall(extract_path)

# Step 2: Load the main CSV file
file_path = os.path.join(extract_path, "API_SP.POP.TOTL_DS2_en_csv_v2_1237901.csv")
df = pd.read_csv(file_path, skiprows=4)

# Step 3: Select data for 2023 and sort by population
pop_2023 = df[["Country Name", "2023"]].dropna()
pop_2023 = pop_2023.sort_values(by="2023", ascending=False).head(10)  # Top 10 countries

# Step 4: Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(pop_2023["Country Name"], pop_2023["2023"], color="pink", edgecolor="black")
plt.title("Top 10 Most Populous Countries (2023)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt


csv_path = None
for root, dirs, files in os.walk("."):  
    for file in files:
        if file.startswith("API_SP.POP.TOTL") and file.endswith(".csv"):
            csv_path = os.path.join(root, file)
            break
    if csv_path:
        break

if csv_path is None:
    raise FileNotFoundError("Population dataset CSV file not found. Please unzip it first.")

print(f"✅ CSV file found at: {csv_path}")

df = pd.read_csv(csv_path, skiprows=4)


pop_values = df["2023"].dropna()


plt.figure(figsize=(10, 6))
plt.hist(pop_values, bins=30, edgecolor='black', color='purple')
plt.title("Distribution of Country Populations (2023)")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.tight_layout()
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt


csv_path = None
for root, dirs, files in os.walk("."):
    for file in files:
        if file.startswith("API_SP.POP.TOTL") and file.endswith(".csv"):
            csv_path = os.path.join(root, file)
            break
    if csv_path:
        break

if csv_path is None:
    raise FileNotFoundError("Population dataset CSV file not found. Please unzip it first.")

print(f"✅ Using file: {csv_path}")

df = pd.read_csv(csv_path, skiprows=4)


countries = ["India", "China", "United States", "Brazil", "Indonesia"]
years = [str(year) for year in range(1960, 2024)]  # Years from 1960–2023


df_filtered = df[df["Country Name"].isin(countries)]
df_filtered = df_filtered.set_index("Country Name")[years].transpose()


plt.figure(figsize=(10, 6))
for country in countries:
    plt.plot(df_filtered.index, df_filtered[country], label=country)

plt.title("Population Growth (1960–2023)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt


csv_path = None
for root, dirs, files in os.walk("."):
    for file in files:
        if file.startswith("API_SP.POP.TOTL") and file.endswith(".csv"):
            csv_path = os.path.join(root, file)
            break
    if csv_path:
        break

if csv_path is None:
    raise FileNotFoundError("Population dataset CSV file not found. Please unzip it first.")

print(f"✅ Using file: {csv_path}")


df = pd.read_csv(csv_path, skiprows=4)


regions = df[df["Country Name"].str.contains("Africa|Asia|Europe|America|Pacific|World", case=False, na=False)]


region_pop = regions[["Country Name", "2023"]].dropna()
region_pop = region_pop.sort_values(by="2023", ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(region_pop["Country Name"], region_pop["2023"], color="cornflowerblue", edgecolor="black")
plt.title("Population by Region (2023)")
plt.xlabel("Region")
plt.ylabel("Population")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
